```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None) #把所有的欄位顯示
```

```python
df = pd.read_csv('data_for_web.csv')
df
```

```python
df['ds'] = df['date'].astype(str)#改成字串型態

df['ds'] = pd.to_datetime(df['ds'],format='%Y%m') #改為字串後，再改成日期格式

df['y'] = df['avg']

df_prophet = df [['ds','y']]

df_prophet =  df_prophet.groupby(['ds',],as_index=True).mean()[['y']].reset_index() #匯整每個月的坪數價格並且求出平均數

df_prophet 
```

```python
model = Prophet(seasonality_mode='multiplicative').fit(df_prophet)  #建模時加入季節變因參數，並且開始訓練模型
    
future = model.make_future_dataframe(periods=12,freq='M')  #建立未來dataframe
    
```

```python
#把台灣國定假日加入變因，驗證對房價是否有影響?

Chinese_New_Year  = pd.DataFrame({
  'holiday':'Chinese New Year\'''s'' Eve',
  'ds': pd.to_datetime(['2013-02-01', '2014-02-01', '2015-02-01',
                        '2016-02-01', '2017-02-01', '2018-02-01',
                        '2019-02-01', '2020-02-01', '2021-02-01',
                        ]),
  'lower_window': -7,
  'upper_window': 7,
})
Dragon_Boat_Festival = pd.DataFrame({
  'holiday': 'Dragon Boat Festival',
  'ds': pd.to_datetime(['2013-06-01', '2014-06-01', '2015-06-01','2016-06-01', '2017-06-01', '2018-06-01', '2019-06-01','2020-06-01', '2021-06-01'  ]),
  'lower_window': 0,
  'upper_window': 7,
})

Spring_Festival = pd.DataFrame({
  'holiday': 'Spring Festival',
  'ds': pd.to_datetime(['2013-04-01', '2014-04-01', '2015-04-01','2016-04-01', '2017-04-01', '2018-04-01', '2019-04-01','2020-04-01', '2021-04-01'  ]),
  'lower_window': 0,
  'upper_window': 3,
})


Mid_Autumn_Festival = pd.DataFrame({
  'holiday': 'Mid Autumn Festival',
  'ds': pd.to_datetime(['2013-09-15', '2014-09-15', '2015-09-15','2016-09-15', '2017-09-15', '2018-09-15', '2019-09-15','2020-09-15', '2021-09-15'  ]),
  'lower_window': 0,
  'upper_window': 3,
})

National_Day = pd.DataFrame({
  'holiday': 'National Day',
  'ds': pd.to_datetime(['2013-10-10', '2014-10-10', '2015-10-10','2016-10-10', '2017-10-10', '2018-10-10', '2019-10-10','2020-10-10', '2021-10-10'  ]),
  'lower_window': 0,
  'upper_window': 3,})


holidays = pd.concat((Chinese_New_Year, Dragon_Boat_Festival,Spring_Festival,Mid_Autumn_Festival,National_Day))
```

```python
model = Prophet(holidays=holidays) #參數加入假日變因
model.add_country_holidays(country_name='Taiwan') #指名台灣
model.fit(df_prophet)

```

```python
forecast = model.predict(future)
fig = model.plot_components(forecast)
fig
```
