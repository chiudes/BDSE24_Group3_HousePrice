```python
import pandas as pd
a = [102,103,104,105,106,107,108,109,110]
pdList=[]
for i in a:
    df = pd.read_csv(f'C:/Users/Student/Desktop/population/pop/ntai_pop/{i}年1月新北市人口統計.csv')
    # rename
    data=df.rename(columns={"隸屬區": "district","合計":"pop","年":"year","戶數":"household"})
    # add cols
    data['sex_ratio']=data['男']/data['女']
    # clean
    data=data.drop([29])
    data=data.drop(["男","女","月", "里數","鄰數"], axis=1)
    data['district']=data['district'].str[0:2]
    pdList.append(data)
print(pdList)

# save file
ntpop=pd.concat(pdList)
ntpop.to_csv('./ntpop.csv',index=False)
```
