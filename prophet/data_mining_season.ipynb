```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None) #把所有的欄位顯示
```

```python
df = pd.read_csv('/share/data/2.1_web.csv')
df
```

```python
df['ds'] = df['date'].astype(str)

df['ds'] = pd.to_datetime(df['ds'],format='%Y%m') #改為字串後，再改成日期格式

df['y'] = df['avg']

df_prophet = df [['ds','y']]

df_prophet =  df_prophet.groupby(['ds',],as_index=True).mean()[['y']].reset_index() #匯整每個月的坪數價格並且求出平均數

df_prophet
```

```python
from fbprophet.plot import plot_yearly
model = Prophet(seasonality_mode='multiplicative').fit(df_prophet)
avg = plot_yearly(m) #已季表示坪數價格波動
```
