```python
df.rename(columns = {'時間':'ROC_year',
                     'M1A原始值':'M1A',
                     'M1A年增率(%)':'M1A_rate',
                     'M1B原始值':'M1B',
                     'M1B年增率(%)':'M1B_rate',
                     'M2原始值':'M2',
                     'M2年增率(%)':'M2_rate',
                    }, inplace = True)
```

```python
df['ROC_year']=df['ROC_year'].str.split("M").str.get(0) + df['ROC_year'].str.split("M").str.get(1)
```

```python
df['ROC_year'] = df['ROC_year'].astype('int')
df['ROC_year'] = df['ROC_year']-191100
```

```python
# 轉換成main dataframe一樣的欄位方便合併
df.rename(columns = {'ROC_year':'date'}, inplace = True)
```

```python
df.to_csv("currency.csv", index=False)
```
