```python
import pandas as pd
import numpy as np
df = pd.read_excel('C:/Users/Student/Desktop/人口結構/人口數、戶數、性別比/台北_人口數、性別比/104年1-12月各區人口數.ods',sheet_name='1月')
```

```python
# 刪除不必要欄位
aaa= df.drop(["Unnamed: 1", "Unnamed: 2"], axis=1)
```

```python
# 整理多餘row
aaa=aaa.drop([0,1])
```

```python
# 增加年份資訊
aaa['年份']=103
```

```python
aaa
```

```python
aaa['性別比']=aaa['男人口數']/aaa['女人口數']
```

```python
aaa=aaa.drop(['男人口數','女人口數'], axis=1)
```

```python
aaa.to_csv("103_pop.csv",index=None)
```
