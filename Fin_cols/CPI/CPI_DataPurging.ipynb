```python
import pandas as pd
```

```python
df = pd.read_xml('CPI.xml')
```

```python
df
```

```python
df.info()
```

```python
df['year']=df['TIME_PERIOD'].str.split("M").str.get(0)
df['month']=df['TIME_PERIOD'].str.split("M").str.get(1)
```

```python
df['year'] = df['year'].astype('int')
df['month'] = df['month'].astype('int')
```

```python
df.info()
```

```python
# 只抓總指數出來
df = df[df['Item'] == '總指數(民國105年=100)']
```

```python
df
```

```python
# 篩選出 2012~2021 年
df = df[ df['year'] >= 2012 ]
df = df[ df['year'] <= 2021 ]
```

```python
df
```

```python
# 刪除不需要的欄位
df = df.drop(['Item', 'TIME_PERIOD','FREQ'], axis=1)
```

```python
df
```

```python
rate = df[df['TYPE'] == '年增率(%)']
CPI = df[df['TYPE'] == '原始值']
```

```python
rate['Item_VALUE'].unique()
# 年增率有負數，代表通貨緊縮
```

```python
CPI['Item_VALUE'].unique()
```

```python
CPI
```

```python
rate
```

```python
# 新增和房價的一樣格式的時間欄位 

CPI['ROC_year'] =  (CPI['year']-1911)*100 + CPI['month']
rate['ROC_year'] =  (rate['year']-1911)*100 + rate['month']

CPI = CPI.drop(['TYPE','year','month'], axis=1)
rate = rate.drop(['TYPE','year','month'], axis=1)
```

```python
CPI.rename(columns = {'Item_VALUE':'CPI'}, inplace = True)

rate.rename(columns = {'Item_VALUE':'CPI_rate'}, inplace = True)
```

```python
CPI
```

```python
rate
```

```python
# merge 方法可以指定欄位合併
CPI_final = pd.merge(CPI, rate, on='ROC_year')
```

```python
CPI_final
```

```python
# 轉換成main dataframe一樣的欄位方便合併
CPI_final.rename(columns = {'ROC_year':'date'}, inplace = True)
```

```python
CPI_final.to_csv("CPI.csv", index=False)
```
