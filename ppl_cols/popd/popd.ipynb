```python
import pandas as pd
import numpy as np
popd_df = pd.read_csv('./popd.csv')
popd_df
```

```python
# 格式清整
popd_df["縣市"]= popd_df["site_id"].apply(lambda x:str(x)[0:3])
```

```python
popd_df = popd_df[popd_df["縣市"].isin(["新北市","臺北市"])]
```

```python
popd_df["區"]= popd_df["site_id"].apply(lambda x:str(x)[3:5])
```

```python
popd_df = popd_df.drop(["people_total","site_id","Unnamed: 0","index"],axis=1)
```
