```python
#GDP = df[df['Item'] == '當期價格(新台幣百萬元)#8.GDP'] 
GDP = df[df['Item'] == '連鎖實質值(2016為參考年_新台幣百萬元)#8.GDP']  # 這才是正確的
```

```python
GDP
```

```python
GDP['year']=GDP['TIME_PERIOD'].str.split("Q").str.get(0)
GDP['quater']=GDP['TIME_PERIOD'].str.split("Q").str.get(1)
```

```python
GDP['year'] = GDP['year'].astype('int')
GDP['quater'] = GDP['quater'].astype('int')
```

```python
GDP.info()
```

```python
GDP = GDP[ GDP['year'] >= 2012 ]
GDP = GDP[ GDP['year'] <= 2021 ]
```

```python
GDP
```

```python
GDP_rate = GDP[GDP['TYPE'] == '年增率(%)']
GDP = GDP[GDP['TYPE'] == '原始值']
```

```python
print(GDP_rate.shape)
print(GDP.shape)
```

```python
GDP_rate = GDP_rate.drop(['TYPE','Item','FREQ'], axis=1)
GDP = GDP.drop(['TYPE','Item','FREQ','quater'], axis=1)
```

```python
GDP.rename(columns = {'Item_VALUE':'GDP'}, inplace = True)

GDP_rate.rename(columns = {'Item_VALUE':'GDP_rate'}, inplace = True)
```

```python
# GDP_rate
```

```python
GDP
```

```python
# print(GDP_rate.shape)
# print(GDP.shape)
```

```python
# merge 方法可以指定欄位合併
GDP_final = pd.merge(GDP, GDP_rate, on='TIME_PERIOD')
```

```python
# GDP_final.shape
```

```python
# GDP_final
```

```python
GDP_final = GDP_final.drop(['TIME_PERIOD','year_x'], axis=1)
GDP_final.rename(columns = {'year_y':'year'}, inplace = True)
```

```python
GDP_final.to_csv("GDP.csv", index=False)
```
