```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None) #把所有的欄位顯示
```

```python
df = pd.read_csv('/share/data/2.1_web.csv')
df
```

```python
df = df[df['district'] == '新北市鶯歌區' ]
df = df.reset_index()
df
```

```python
df['ds'] = df['date'].astype(str)

df['ds'] = pd.to_datetime(df['ds'],format='%Y%m') #改為字串後，再改成日期格式

df['y'] = df['avg']

df_prophet = df [['ds','y']]

df_prophet =  df_prophet.groupby(['ds',],as_index=True).mean()[['y']].reset_index()

df_prophet 
```

```python
model = Prophet(seasonality_mode='multiplicative').fit(df_prophet)
    
future = model.make_future_dataframe(periods=12,freq='M')  
    
fcst = model.predict(future) 
   
fig = model.plot(fcst)     
    
fig

plt.savefig('新北市鶯歌區')
```

```python
trendFig = model.plot_components(fcst)
    
trendFig 
       
plt.savefig('新北市鶯歌區_year')
```
