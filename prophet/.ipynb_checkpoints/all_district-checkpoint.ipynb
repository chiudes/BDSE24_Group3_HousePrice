```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns", None) #把所有的欄位顯示
```

```python
df = pd.read_csv('data_for_web.csv') #讀取dataframe
df
```

```python
df['ds'] = df['date'].astype(str) #改成字串型態

df['ds'] = pd.to_datetime(df['ds'],format='%Y%m') #改為字串後，再改成日期格式

df['y'] = df['avg']

df_prophet = df [['ds','y']]

df_prophet =  df_prophet.groupby(['ds',],as_index=True).mean()[['y']].reset_index() #匯整每個月的坪數價格並且求出平均數

df_prophet 
```

```python
model = Prophet(seasonality_mode='multiplicative').fit(df_prophet)  #建模
    
future = model.make_future_dataframe(periods=12,freq='M')  #建置未來時間的dataframe，月為單位，且預測期間為1年
    
fcst = model.predict(future) 
   
fig = model.plot(fcst)     
    
fig
```

```python
trendFig = model.plot_components(fcst) #得出未來一年是漲是跌 並且得出2013~2022價格漲幅    
trendFig 
       
```
