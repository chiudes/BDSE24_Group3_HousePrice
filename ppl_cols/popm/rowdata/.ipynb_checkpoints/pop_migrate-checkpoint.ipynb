```python
# excle to csv
import pandas as pd 
import glob
filename=glob.glob('C:/Users/Student/Desktop/人口結構/遷出遷入/台北_遷出遷入/*.ods')
for i in filename:
    df=pd.read_excel(f'{i}')
    df.to_csv('./'+f'{i}'[-9:-4]+'.csv')
    print(i,'done')
```

```python
# clean data
import numpy as np
import pandas as pd
import glob
filename=glob.glob('C:/Users/Student/雙北人口/row data/taipei/*.csv')
for f in filename:
    aaa=pd.read_csv(f)

    # cover data
    aaa=aaa.iloc[0:15:,[1,2]]
    
    # drop cols
    #aaa=aaa.drop(['Unnamed: 0','Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10','Unnamed: 11','Unnamed: 12','Unnamed: 13'],axis=1)
    
    # drop rows
    aaa=aaa.drop([0,1,2])
    
    # rename
    col1=str(aaa.columns[0][8:11])
    popm=str(aaa.columns[0][-5:-3])
    year=str(aaa.columns[0][3:6])
    aaa.rename(columns ={aaa.columns[0]:col1,aaa.columns[1]:popm},inplace=True)
    #aaa['行政區']=aaa['行政區'][:12]
    
    # add col
    aaa=aaa.assign(年份=year)
    
    # save
    aaa.to_csv(f'../{year}{popm}.csv')
    print(f,'done')
```

```python
# concat
import glob
final = pd.concat([pd.read_csv(i) for i in glob.glob('../*.csv')])
final=final.reset_index()
print(final)
final.to_csv('../popm.csv')
```
