```python
import pandas as pd
import glob
from functools import reduce
```

```python
lst=glob.glob('C:/Users/Student/BDSE_pop/cleaned_data/*.csv')
lst
```

```python
pdList=[]
for f in lst:
    data=pd.read_csv(f)
    pdList.append(data)
print(pdList)
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=["year","district"],
                                            how='outer'), pdList)
```

```python
df_merged
```

```python
# add cols 
df_merged['family_pop']=df_merged['pop']/df_merged['household']
df_merged['social_rate']=df_merged['in']-df_merged['out']
df_merged
```

```python
df_merged.to_csv('../popall.csv',index=False)
```
