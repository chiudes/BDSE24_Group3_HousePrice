# merge

```python
df = pd.merge(bank_interestrate, CCI, on=['date'])
df = pd.merge(df, CPI, on=['date'])
df = pd.merge(df, currency, on=['date'])
df = pd.merge(df, load_house_archi, on=['date'])
df = pd.merge(df, Non_Performing_Loans_Ratio, on=['date'])
df = pd.merge(df, TWSE, on=['date'])
df = pd.merge(df, unemployment_rate, on=['date'])
df
```

```python
# 產生 quater 欄位
df['date'] = df['date'].astype('str')
df['month'] = df['date'].str[-2:].astype('int') 

df['quater'] = (df['month'] + 2) // 3 

df['date'] = df['date'].str[:3].astype('int')
```

```python
df
```

```python
df.info()
```

```python
# income 不是每月提供資料，額外處理
df = pd.merge(df, income, on=['date'])
df
```

```python
# 和 GDP 不是每月提供資料，額外處理
df = pd.merge(df, GDP, on=['quater', 'date'])
df
```

```python
df['date'] = df['date'] *100 + df['month']
df = df.drop(['month', 'year', 'quater'], axis=1)
```

```python
df.to_csv("AllFinancialCols.csv", index=False)
```
